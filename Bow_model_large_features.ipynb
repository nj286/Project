```
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix # Confusion matrix
from sklearn.metrics import classification_report # precision, recall, f1

#from sklearn.metrics import f1_score
#from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Capstone_project/cleaned_reviews.csv')
df.head()
```

# As the data is not balanced, considering only 27398 samples for each class.



```
def get_index(x):
  
  index_1, index_2, index_3, index_4, index_5 = [], [], [], [], []
  for ind,val in enumerate(x):

    if val == 1:
      index_1.append(ind)
    elif val == 2:
      index_2.append(ind)
    elif val == 3:
      index_3.append(ind)
    elif val == 4:
      index_4.append(ind)
    else:
      index_5.append(ind)
  print(len(index_1),len(index_2), len(index_3), len(index_4), len(index_5))

  index = index_1[:27398] + index_2[:27398] + index_3[:27398] + index_4[:27398] + index_5[:27398]
  
  print(len(index))
  return index
```

```
index = get_index(df['rating'])
```

```
df = df.loc[index]
```

```
x = df.cleaned_reviews
y = df.rating
```

```
SEED = 2000
x_train, x_validation_and_test, y_train, y_validation_and_test = train_test_split(x, y, test_size=.25, random_state=SEED)
x_validation, x_test, y_validation, y_test = train_test_split(x_validation_and_test, y_validation_and_test, test_size=.5, random_state=SEED)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tvec1 = TfidfVectorizer(max_features=100000,ngram_range=(1, 3))
tvec1.fit(x_train)
```

```
x_train_tfidf = tvec1.transform(x_train)
```

```
x_validation_tfidf = tvec1.transform(x_validation)
```

```
x_test_tfidf = tvec1.transform(x_test)
```

```
# Naive Bayes 
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()

# Training the model on the udersampled data.

clf.fit(x_train_tfidf, y_train)
# Prediction on the 
pred = clf.predict(x_validation_tfidf)
# Confusion Matrix
print(confusion_matrix(y_validation, pred))

# Precision, Recall, F1-score 
print(classification_report(y_validation, pred))
```

```
pred_nb = clf.predict(x_test_tfidf)

print(confusion_matrix(y_test, pred_nb))

# Precision, Recall, F1-score 
print(classification_report(y_test, pred_nb))
```
